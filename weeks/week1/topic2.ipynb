## Question 1.4:  What's the rounded difference between median values of age (in months) for non-smokers and smokers? You'll need to figure out the units of feature `age` in this dataset.

```python
median_nonsmokers_days = df[df['smoke'] == 0]['age'].median()
median_smokers_days = df[df['smoke'] == 1]['age'].median()

print(f"Median age non-smokers: {median_nonsmokers_days} days")
print(f"Median age smokers: {median_smokers_days} days")

median_nonsmokers_months = median_nonsmokers_days / 365.25 * 12
median_smokers_months = median_smokers_days / 365.25 * 12

print(f"Median age non-smokers: {median_nonsmokers_months:.1f} months")
print(f"Median age smokers: {median_smokers_months:.1f} months")

diff_months = abs(median_nonsmokers_months - median_smokers_months)
print(f"Difference: {diff_months:.1f} months")
```

## Question 1.5: Calculate fractions of ill people (with CVD) in the two groups of people described in the task. What's the ratio of these two fractions?


```python
# create age in years column
df['age_years'] = (df['age'] / 365.25).round().astype(int)

target_group = df[
    (df['gender'] == 2) & 
    (df['age_years'] >= 60) & (df['age_years'] <= 64) & 
    (df['smoke'] == 1)
]

print(f"Target group size: {len(target_group)}")
```

```python
# Group 1: systolic BP < 120 and cholesterol = 1
group1 = target_group[
    (target_group['ap_hi'] < 120) & 
    (target_group['cholesterol'] == 1)
]

print(f"Group 1 size: {len(group1)}")
print(f"Group 1 with CVD: {group1['cardio'].sum()}")
fraction1 = group1['cardio'].mean() if len(group1) > 0 else 0
print(f"Group 1 CVD fraction: {fraction1:.4f}")
```

```python
# Group 2: systolic BP 160-179 and cholesterol = 3
group2 = target_group[
    (target_group['ap_hi'] >= 160) & 
    (target_group['ap_hi'] < 180) & 
    (target_group['cholesterol'] == 3)
]

print(f"Group 2 size: {len(group2)}")
print(f"Group 2 with CVD: {group2['cardio'].sum()}")
fraction2 = group2['cardio'].mean() if len(group2) > 0 else 0
print(f"Group 2 CVD fraction: {fraction2:.4f}")
```

```python
# Calculate ratio
ratio = fraction2 / fraction1 if fraction1 > 0 else 0
print(f"Ratio (Group 2 / Group 1): {ratio:.2f}")
print(f"Rounded ratio: {round(ratio)}")
```

## Question 1.6: Choose the correct statements about BMI

```python
# Calculate BMI
df['BMI'] = df['weight'] / (df['height'] / 100) ** 2

print("BMI Statistics:")
print(df['BMI'].describe())
```

```python
# Statement 1: Median BMI in the sample is within boundaries of normal values
median_bmi = df['BMI'].median()
print(f"Statement 1: Median BMI in normal range (18.5-25)?")
print(f"Median BMI: {median_bmi:.2f}")
print(f"Result: {18.5 <= median_bmi <= 25}")
```

```python
# Statement 2: Women's BMI is on average higher than men's
bmi_women = df[df['gender'] == 1]['BMI'].mean()
bmi_men = df[df['gender'] == 2]['BMI'].mean()

print(f"Statement 2: Women's BMI > Men's BMI?")
print(f"Women BMI: {bmi_women:.2f}")
print(f"Men BMI: {bmi_men:.2f}")
print(f"Result: {bmi_women > bmi_men}")
```

```python
# Statement 3: Healthy people have higher median BMI than ill people
bmi_healthy = df[df['cardio'] == 0]['BMI'].median()
bmi_ill = df[df['cardio'] == 1]['BMI'].median()

print(f"Statement 3: Healthy median BMI > Ill median BMI?")
print(f"Healthy median BMI: {bmi_healthy:.2f}")
print(f"Ill median BMI: {bmi_ill:.2f}")
print(f"Result: {bmi_healthy > bmi_ill}")
```

```python
# Statement 4: In the segment of healthy and non-drinking men, BMI is closer to norm than in women
healthy_nondrinking_men = df[(df['cardio'] == 0) & (df['alco'] == 0) & (df['gender'] == 2)]
healthy_nondrinking_women = df[(df['cardio'] == 0) & (df['alco'] == 0) & (df['gender'] == 1)]

bmi_men_healthy = healthy_nondrinking_men['BMI'].mean()
bmi_women_healthy = healthy_nondrinking_women['BMI'].mean()
bmi_norm_center = 21.75  # (18.5 + 25) / 2

print(f"Statement 4: Healthy non-drinking men BMI closer to norm?")
print(f"Men BMI: {bmi_men_healthy:.2f}")
print(f"Women BMI: {bmi_women_healthy:.2f}")
print(f"Norm center: {bmi_norm_center}")
print(f"Men distance from norm: {abs(bmi_men_healthy - bmi_norm_center):.2f}")
print(f"Women distance from norm: {abs(bmi_women_healthy - bmi_norm_center):.2f}")
print(f"Result: {abs(bmi_men_healthy - bmi_norm_center) < abs(bmi_women_healthy - bmi_norm_center)}")
```

## Question 1.7: What percent of the original data (rounded) did we filter out in the previous step?

```python
original_count = len(df)
print(f"Original dataset size: {original_count}")

df_to_remove = df[
    (df['ap_lo'] > df['ap_hi']) |
    (df['height'] < df['height'].quantile(0.025)) |
    (df['height'] > df['height'].quantile(0.975)) |
    (df['weight'] < df['weight'].quantile(0.025)) |
    (df['weight'] > df['weight'].quantile(0.975))
]

removed_count = len(df_to_remove)
print(f"Removed count: {removed_count}")

percent_removed = removed_count / original_count * 100
print(f"Percentage removed: {percent_removed:.2f}%")
```

```python
# new clean dataset
df_clean = df[~df.index.isin(df_to_remove.index)].copy()
print(f"Clean dataset size: {len(df_clean)}")
```

## Question 2.2: Which pair of features has the strongest Spearman rank correlation?

```python
spearman_corr = df_clean.corr(method='spearman')

corr_pairs = []
for i in range(len(spearman_corr.columns)):
    for j in range(i+1, len(spearman_corr.columns)):
        col1 = spearman_corr.columns[i]
        col2 = spearman_corr.columns[j]
        corr_val = abs(spearman_corr.iloc[i, j])
        corr_pairs.append((col1, col2, corr_val))

corr_pairs_sorted = sorted(corr_pairs, key=lambda x: x[2], reverse=True)

print("Top 10 Spearman correlations:")
for col1, col2, val in corr_pairs_sorted[:10]:
    print(f"{col1:15s} - {col2:15s}: {val:.4f}")
```

## Question 2.4: What is the smallest age at which the number of people with CVD outnumbers the number of people without CVD?

```python
cvd_by_age = df_clean.groupby(['age_years', 'cardio']).size().unstack(fill_value=0)
cvd_by_age.columns = ['No CVD', 'CVD']

print("CVD counts by age:")
print(cvd_by_age.tail(20))
```

```python
age_threshold = cvd_by_age[cvd_by_age['CVD'] > cvd_by_age['No CVD']].index.min()
print(f"Age threshold where CVD > No CVD: {age_threshold}")
```
