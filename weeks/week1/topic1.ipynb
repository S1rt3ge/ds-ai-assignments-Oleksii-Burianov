```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

pd.set_option('display.max_columns', 100)
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
DATA_URL = 'https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/'
data = pd.read_csv(DATA_URL + 'adult.data.csv')

print(f"Dataset shape: {data.shape}")
data.head()
```

### Question 1: How many men and women (sex feature) are represented in this dataset?

```python
sex_counts = data['sex'].value_counts()
print("Gender distribution:")
print(sex_counts)
print(f"\nMen: {sex_counts['Male']}")
print(f"Women: {sex_counts['Female']}")
print(f"\nPercentage of men: {sex_counts['Male'] / len(data) * 100:.1f}%")
print(f"Percentage of women: {sex_counts['Female'] / len(data) * 100:.1f}%")
```

### Question 2: What is the average age (age feature) of women?

```python
avg_age_women = data[data['sex'] == 'Female']['age'].mean()
print(f"Average age of women: {avg_age_women:.2f} years")
```

### Question 3: What is the percentage of German citizens (native-country feature)?

```python
german_count = (data['native-country'] == 'Germany').sum()
german_percentage = german_count / len(data) * 100

print(f"Number of Germans: {german_count}")
print(f"Percentage of German citizens: {german_percentage:.2f}%")
```

### Question 4: What are the mean and standard deviation of age for those who earn more than 50K per year?

```python
ages_rich = data[data['salary'] == '>50K']['age']

mean_age_rich = ages_rich.mean()
std_age_rich = ages_rich.std()

print(f"For those earning >50K:")
print(f"  Mean age: {mean_age_rich:.1f} years")
print(f"  Standard deviation: {std_age_rich:.1f} years")
```

### Question 5: What are the mean and standard deviation of age for those who earn less than 50K per year?

```python
ages_poor = data[data['salary'] == '<=50K']['age']

mean_age_poor = ages_poor.mean()
std_age_poor = ages_poor.std()

print(f"For those earning <=50K:")
print(f"  Mean age: {mean_age_poor:.1f} years")
print(f"  Standard deviation: {std_age_poor:.1f} years")
```

### Question 6: Is it true that people who earn more than 50K have at least high school education?

```python
high_education = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
rich_people = data[data['salary'] == '>50K']

print("Unique education levels for people earning >50K:")
print(rich_people['education'].unique())
print()

# check if there are people without high school education
low_education = rich_people[~rich_people['education'].isin(high_education + ['HS-grad', 'Some-college'])]
print(f"Number of high earners WITHOUT high school education: {len(low_education)}")
print(f"These include: {low_education['education'].unique()}")
```

### Question 7: Find the maximum age of men of Amer-Indian-Eskimo race.

```python
# age statistics by race and gender
age_stats = data.groupby(['race', 'sex'])['age'].describe()
print("Age statistics by race and gender:")
print(age_stats)

# maximum age of Amer-Indian-Eskimo men
max_age = data[(data['race'] == 'Amer-Indian-Eskimo') & 
               (data['sex'] == 'Male')]['age'].max()
print(f"Maximum age of Amer-Indian-Eskimo men: {max_age} years")
```

### Question 8: Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)? Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.?

```python
# married men
married_men = data[(data['sex'] == 'Male') & 
                   (data['marital-status'].str.startswith('Married'))]
married_rich_prop = (married_men['salary'] == '>50K').sum() / len(married_men)

# single men (never-married)
single_men = data[(data['sex'] == 'Male') & 
                  (data['marital-status'] == 'Never-married')]
single_rich_prop = (single_men['salary'] == '>50K').sum() / len(single_men)

print("Proportion of high earners (>50K):")
print(f"Married men: {married_rich_prop:.2%} ({(married_men['salary'] == '>50K').sum()} out of {len(married_men)})")
print(f"Single men:  {single_rich_prop:.2%} ({(single_men['salary'] == '>50K').sum()} out of {len(single_men)})")
print(f"Difference: {(married_rich_prop - single_rich_prop) * 100:.1f} percentage points")
```

### Question 9: What is the maximum number of hours a person works per week? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?

```python
max_hours = data['hours-per-week'].max()
num_workaholics = (data['hours-per-week'] == max_hours).sum()
workaholics = data[data['hours-per-week'] == max_hours]
rich_workaholics = (workaholics['salary'] == '>50K').sum()
percentage = rich_workaholics / len(workaholics) * 100

print(f"Maximum hours per week: {max_hours} hours")
print(f"Number of people working {max_hours} hours/week: {num_workaholics}")
print(f"Earning >50K among them: {rich_workaholics} out of {num_workaholics}")
print(f"Percentage earning >50K: {percentage:.1f}%")
```

### Question 10: Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country). What will these be for Japan?

```python
# average hours by country and salary
avg_hours_by_country_salary = data.groupby(['native-country', 'salary'])['hours-per-week'].mean()

# Japan specifically
print("Average hours per week by country and salary level:")
print("=" * 60)
print("\nFor Japan:")
japan_data = avg_hours_by_country_salary['Japan']
print(f"  Earning <=50K: {japan_data['<=50K']:.1f} hours/week")
print(f"  Earning >50K:  {japan_data['>50K']:.1f} hours/week")

# top 10 countries for reference
print("\n" + "=" * 60)
print("Top 10 countries by sample size:")
top_countries = data['native-country'].value_counts().head(10).index
for country in top_countries:
    if country in avg_hours_by_country_salary.index:
        low = avg_hours_by_country_salary[country]['<=50K']
        high = avg_hours_by_country_salary[country]['>50K']
        print(f"{country:20s}: <=50K: {low:.1f}h, >50K: {high:.1f}h")
```
